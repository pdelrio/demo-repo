1. Write a Python program to obtain the frequency of words from the given file. (sample .txt)

```python
from google.colab import files #import file upload
files.upload() #choose file to upload 'sample .txt'
```

```python

def word_freq(file_name):
  """
  Given a file find the word count
  """
  freq = {} #initialize an empty dictionary to store frequencies
  with open(file_name, 'r') as file: #open file using with

    for line in file:
      words = line.split() #for each line it splits the line into words and iterates to each word
      for word in words:
        word = ''.join(filter(str.isalnum, word)) #for each word removes punctuation
        word = word.lower() #make each word into lowercases
        if word:
          freq[word] = freq.get(word, 0) + 1 #adds word to frequency

  return freq

file_name = 'sample .txt'

result = word_freq(file_name) #call function with file


for word, frequency in result.items():
  print(f"{word}: {frequency}") #print dictionary output

```

2. Write a function in Python to count words in a text file ending with alphabet "e"

```python
import re #import regex

def count_words_ending_with_e(file):
    """
    Count words in a text file ending with "e"
    """

    with open(file, 'r') as file: #open the file and read content
        text = file.read()

    words = re.findall(r'\w+', text) #split the text into words

    words_ending_with_e = [word for word in words if word.lower().endswith('e')] #filter lowercased words ending with e

    word_freq = {} #Count frequency of words ending with'e'
    for word in words_ending_with_e:
        word_freq[word] = word_freq.get(word, 0) + 1

    # Print the word frequencies
    for word, count in word_freq.items():
        print(f"{word}: {count}") #print frequencies

    print(f"No of words that end with 'e' are {len(word_freq)}") #print output according to assignment guideline

count_words_ending_with_e("sample .txt") #call function
```

```python
from pickle import TRUE
import pandas as pd
data_2 = pd.read_csv('drinks.csv', keep_default_na= False) #after the initial inspection, I realized all North American countries became NaN so I upoload the file again removin NA as NaN by default

#print(data_2.to_string()) #used this to inspect the data and make decisions how and what clean/change
#print(data_2.info()) #used to inspect data


#CLEAN DATA
#The first issue with North America becoming NA was fixed when importing the file
#There are 193 countries same as the ones members of the United Nations so it does not seem like duplicates no duplicates to drop
#There are some countries with alcohol conmsuption of zero but this is explained but legal and religious ressons


# a) WHAT IS THE AVERAGE OF BEER CONSUMPTION PER CONTINENT?

grouped_continent = data_2.groupby('continent')['beer_servings'].mean().reset_index() #grouped data by continets, reset_index() conver result of mean() to a DataFrame

for index, row in grouped_continent.iterrows(): #iterate over the rows of the grouped DataFrame using iterrows()
  continent = row['continent']
  avg_beer = row['beer_servings']
  print(f"Average of beer consumption in {continent}: {avg_beer:.2f}") #print output

print("---------------------------------------------------------------------------------------")

# b) PRINT THE MEAN ALCOHOL CONSUMPTION PER CONTINENT FOR EVERY COLUMN
drinks = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol'] #col of new Data Frame
means_per_continent = data_2.groupby(['continent'])[drinks].mean().reset_index() #create a new DataFrame with average of alchol consumption per continent

print(means_per_continent.to_string())

print("---------------------------------------------------------------------------------------")

# c) WHICH CONTINENT CONSUMES THE LEAST AMOUNT OF ALCOHOL PER CAPITA?

means_per_continent['total'] = means_per_continent[drinks].sum(axis=1) #using the DataFrame for the question b), add another column for total averages of all beverages
less_alcohol = means_per_continent.min()['total'] #find the minimun total average
less_alcohol_index = means_per_continent['total'].idxmin() #find the index of the minimun total average
cont_name_less_alco = means_per_continent.loc[less_alcohol_index, 'continent'] #find the continent the min total average belong to
print(f"The continent that consumes the less alcohol is {cont_name_less_alco} with a total average average percapita consumption of {less_alcohol:.2f} ") #print output

print("----------------------------------------------------------------------------------------")

# d) WHAT IS THE TOTAL CONSUMPTION OF PURE ALCOHOL BY CONTINENT?

grouped_alcohol = data_2.groupby('continent')['total_litres_of_pure_alcohol'].sum().reset_index() #gropuped data by continents, add total litres of alc, reset_index to convert result of sum to a DataFrame

for index, row in grouped_alcohol.iterrows(): #iterate over the rows fo the grouped DataFrame
  continental = row['continent']
  total_pure_al = row['total_litres_of_pure_alcohol']
  print(f"The total pure alcohol consumption by {continental}: {total_pure_al} ") #print output

print("------------------------------------------------------------------------------------------")

# c) WHAT IS THE COUNTRY WITH THE HIGHEST TOTAL ALCOHOL CONSUMPTION

data_2['total'] = data_2[drinks].sum(axis=1) #add a total column to the data frame that stores the total of all beverages
country_more = data_2.max()['total'] #find the max of total
country_more_index = data_2['total'].idxmax()#find the index fo the max total
name_country_more = data_2.loc[country_more_index, 'country'] #find the continent that the max total index belongs
print(f"The country with the highest total alcohol consumption is {name_country_more} with a total consumption of {country_more}") #print output
```
